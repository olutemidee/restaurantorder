```python
import sqlite3
import pandas as pd
import numpy as np
df_Orders = pd.read_csv("Orders.csv")



def replace_na_characters(df_Orders, threshold=0.5):
    # Replace 'NA' in columns with characters and many missing values
    for column in df_Orders.columns:
        if df_Orders[column].dtype == 'O' and df_Orders[column].isna().mean() > threshold:
            df_Orders[column].fillna(value='N/A', inplace=True)

def fill_na_numbers(df_Orders, threshold=0.1):
    # Fill 'NA' in columns with numbers and few missing values
    for column in df_Orders.columns:
        if np.issubdtype(df_Orders[column].dtype, np.number) and df_Orders[column].isna().mean() <= threshold:
            mean_value = df_Orders[column].mean()
            df_Orders[column].fillna(value=mean_value, inplace=True)

def treat_credit_debit_cards(df_Orders):
    # Treat 'Credit Card' and 'Debit Card'and Card Provider columns separately
    df_Orders['Credit_Card'].fillna(value='No Card', inplace=True)
    df_Orders['Debit_Card'].fillna(value='No Card', inplace=True)
    df_Orders['Card_provider'].fillna(value='N/A', inplace=True)

def main():
    # Read your CSV file into a pandas DataFrame
    csv_file_path = 'Orders.csv'
    df_Orders = pd.read_csv(csv_file_path)

    # Display the original DataFrame
    print("Original DataFrame:")
    print(df_Orders)

    # Replace 'NA' in columns with characters and many missing values
    replace_na_characters(df_Orders)

    # Fill 'NA' in columns with numbers and few missing values
    fill_na_numbers(df_Orders)

    # Treat 'Credit Card' and 'Debit Card' columns separately
    treat_credit_debit_cards(df_Orders)

    # Display the cleaned DataFrame
    print("\nCleaned DataFrame:")
    print(df_Orders)

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import messagebox
import sqlite3

class MeanRatingCalculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Mean Rating Calculator")

        self.label_result = tk.Label(root, text="Mean Customer Rating-Food:")
        self.label_result.pack()

        self.button_calculate_mean = tk.Button(root, text="Calculate Mean", command=self.calculate_mean)
        self.button_calculate_mean.pack()

    def calculate_mean(self):
        try:
            connection = sqlite3.connect("takefood.db")
            cursor = connection.cursor()

            # Execute the SQL query to calculate the mean Customer Rating-Food for each restaurant
            cursor.execute("""
                SELECT Restaurant_ID, AVG("Customer Rating-Food") AS Mean_Rating_Food
                FROM Orders
                GROUP BY Restaurant_ID
            """)

            results = cursor.fetchall()

            connection.close()

            if results:
                # Display the calculated mean ratings
                result_message = ""
                for result in results:
                    restaurant_id, mean_rating_food = result
                    result_message += f"Restaurant ID {restaurant_id}: Mean Rating - Food: {mean_rating_food:.2f}\n"

                self.label_result.config(text=result_message)
            else:
                messagebox.showinfo("No Data", "No data found in the orders table.")

        except sqlite3.Error as e:
            messagebox.showerror("Database Error", f"Error calculating mean ratings: {str(e)}")

if __name__ == "__main__":
    root = tk.Tk()
    app = MeanRatingCalculator(root)
    root.mainloop()

```

```python
import tkinter as tk
from tkinter import messagebox
import sqlite3
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

class HistogramDrawer:
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Time Histogram")

        self.button_draw_histogram = tk.Button(root, text="Draw Histogram", command=self.draw_histogram)
        self.button_draw_histogram.pack()

    def draw_histogram(self):
        try:
            connection = sqlite3.connect("takefood.db")
            cursor = connection.cursor()

            # Execute the SQL query to fetch delivery times
            cursor.execute("SELECT [Delivery_Time _Taken(mins)] FROM Orders")
            delivery_times = cursor.fetchall()

            connection.close()

            if delivery_times:
                # Flatten the list of delivery times and take absolute value
                delivery_times = [abs(time[0]) for time in delivery_times]

                # Draw the histogram
                fig, ax = plt.subplots()
                ax.hist(delivery_times, bins=20, color='blue', edgecolor='black')
                ax.set_xlabel('Delivery Time Taken (mins)')
                ax.set_ylabel('Frequency')
                ax.set_title('Histogram of Delivery Time Taken')

                # Embed the plot in the Tkinter window
                canvas = FigureCanvasTkAgg(fig, master=self.root)
                canvas.draw()
                canvas.get_tk_widget().pack()

                # Show the Tkinter window
                plt.show()

            else:
                messagebox.showinfo("No Data", "No delivery time data found in the Orders table.")

        except sqlite3.Error as e:
            messagebox.showerror("Database Error", f"Error fetching delivery times: {str(e)}")

if __name__ == "__main__":
    root = tk.Tk()
    app = HistogramDrawer(root)
    root.mainloop()

   
```

```python
import tkinter as tk
from tkinter import messagebox
import sqlite3
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import statistics

class StatisticsViewer:
    def __init__(self, root):
        self.root = root
        self.root.title("Statistics Viewer")

        self.button_show_statistics = tk.Button(root, text="Show Statistics", command=self.show_statistics)
        self.button_show_statistics.pack()

    def show_statistics(self):
        try:
            connection = sqlite3.connect("takefood.db")
            cursor = connection.cursor()

            # Execute the SQL query to fetch delivery times
            cursor.execute("SELECT [Delivery_Time _Taken(mins)] FROM Orders")
            delivery_times = cursor.fetchall()

            connection.close()

            if delivery_times:
                # Flatten the list of delivery times
                delivery_times = [time[0] for time in delivery_times]

                # Calculate statistics
                mean_time = statistics.mean(delivery_times)
                median_time = statistics.median(delivery_times)
                stdev_time = statistics.stdev(delivery_times)

                # Display the statistics
                statistics_message = (
                    f"Mean Delivery Time: {mean_time:.2f} mins\n"
                    f"Median Delivery Time: {median_time:.2f} mins\n"
                    f"Standard Deviation: {stdev_time:.2f} mins"
                )
                messagebox.showinfo("Statistics", statistics_message)

            else:
                messagebox.showinfo("No Data", "No delivery time data found in the Orders table.")

        except sqlite3.Error as e:
            messagebox.showerror("Database Error", f"Error fetching delivery times: {str(e)}")

if __name__ == "__main__":
    root = tk.Tk()

    # Create instances of both HistogramDrawer and StatisticsViewer
    app_histogram = HistogramDrawer(root)
    app_statistics = StatisticsViewer(root)

    root.mainloop()
```

```python
import sqlite3

connection = sqlite3.connect("takefood.db")
cursor = connection.cursor()

try:
    # Drop the existing delivery_staff table if it exists
    cursor.execute("DROP TABLE IF EXISTS delivery_staff;")
    
    # Create the delivery_staff table
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS delivery_staff (
            Delivery_id INTEGER PRIMARY KEY,
            Order_ID INTEGER,
            name TEXT,
            vehicle_types TEXT,
            plate_number TEXT,
            contact_number TEXT
        );
    """)
    
    connection.commit()
    print("Created delivery_staff table successfully.")
except sqlite3.Error as e:
    print(f"Error creating table: {str(e)}")
finally:
    # Close the database connection
    connection.close()
```
